```python
import requests
from bs4 import BeautifulSoup
import yfinance as yf
from datetime import datetime
from datetime import timedelta
import pandas as pd
import re
import json
import dateutil.parser
from selenium import webdriver
import time
from functions import (get_sector, changes_from_press, get_content_first_ele_lmt,
                       get_df, normalise_to_index, save_data)
from requests_html import HTMLSession
import nest_asyncio
from selenium import webdriver

nest_asyncio.apply()
past_last_date = False
```

```python
data = []
offset = 0

ticker = 'lmt'

sector = get_sector(ticker)

index_ticker = '^GSPC'
# select by value 

stock_hist = get_df(ticker)
index_hist = get_df(index_ticker)


while True:
    
    #headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/94.0.4606.81 Safari/537.36"}
    
   
    html = requests.get(f'https://news.lockheedmartin.com/news-releases?category=788&l=20&o={offset}', headers=headers)

    soup = BeautifulSoup(html.content) 

    articles = soup.find_all('li', {'class':'wd_item'})
    for article in articles:
        date = article.find('div', attrs={'class':'wd_date'}).text.lstrip().rstrip() 
        title = article.find('a').text.lstrip().rstrip() 
        
        url = article.find('a').get('href')
        
        date = date.replace(',', '')    
        date = dateutil.parser.parse(date).date()
        
        if url == 'https://news.lockheedmartin.com/2016-10-17-Lockheed-Martin-Celebrates-25-years-of-Work-on-the-Royal-Navys-Merlin-Helicopter-Programme':
            continue

        print(url)
        
        content = get_content_first_ele_lmt(url, 'p', {})
        
        if date < datetime(2014, 1, 1).date():
            past_last_date = True
            break        
    
        pct_change = changes_from_press(stock_hist, date, 1)
        
        if pct_change != None:
            index_price, norm_price = normalise_to_index(pct_change, date, index_hist)
        else:
            index_price, norm_price = (None, None)

        data.append([date, title, url, content, pct_change, index_price, norm_price])
    if past_last_date == True: break    
    offset = offset + 20
    
save_data(data, ticker, sector)

past_last_date = False
```
