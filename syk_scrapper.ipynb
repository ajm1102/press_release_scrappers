```python
import requests
from bs4 import BeautifulSoup
import yfinance as yf
from datetime import datetime
from datetime import timedelta
import pandas as pd
from selenium import webdriver
import dateutil.parser
import re
import json
import time
from functions import (get_sector, changes_from_press, get_content_header,
                       get_df, normalise_to_index, save_data)

past_last_date = False
```

```python
data = []
year = datetime.now().year

ticker = 'syk'

sector = get_sector(ticker)

index_ticker = '^GSPC'
# select by value 

stock_hist = get_df(ticker)
index_hist = get_df(index_ticker)


while True:
    

    driver = webdriver.Firefox()
    driver.get(f'https://investors.stryker.com/press-releases/default.aspx#{year}/')

    time.sleep(2)
    html = driver.page_source
    soup = BeautifulSoup(html) 

    articles = soup.find_all('div', {'class':'module_item'})
    for article in articles:
        
        date = article.find('span', attrs={'class':'module_date-text'}).text.lstrip().rstrip() 
        title = article.find('a', attrs={'class':'module_headline-link'}).text.lstrip().rstrip() 
        url = article.find('a', attrs={'class':'module_headline-link'}).get('href') 
        print(date)
        url = f'https://investors.stryker.com{url}'
        
        date = dateutil.parser.parse(date).date()

        content = get_content_header(url, 'meta' , {'name': 'twitter:description'})
        print(url)
        if date < datetime(2014, 1, 1).date():
            past_last_date = True
            break        
    
        pct_change = changes_from_press(stock_hist, date, 1)
        
        if pct_change != None:
            index_price, norm_price = normalise_to_index(pct_change, date, index_hist)
        else:
            index_price, norm_price = (None, None)

        data.append([str(date), title, url, content, pct_change, index_price, norm_price])
    if past_last_date == True: break    
    year = year - 1
    
save_data(data, ticker, sector)

past_last_date = False
```
