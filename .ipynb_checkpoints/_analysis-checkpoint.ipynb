```python
import os, json
import pandas as pd
import numpy as np
import seaborn as sns
import glob
import datetime
import matplotlib.pyplot as plt
from dateutil.parser import parse

press_data = []
norm_prices = []
pct_changes = []

path_to_json = './data/press_data/'

for file_name in [file for file in os.listdir(path_to_json) if file.endswith('.json')]:
    with open(path_to_json + file_name, encoding="utf8") as json_file:
        file = json.load(json_file)
        press_data.extend(file)
```

```python
daily_data = []
path_to_json = './data/all_daily/'

for file_name in [file for file in os.listdir(path_to_json) if file.endswith('.json')]:
    with open(path_to_json + file_name, encoding="utf8") as json_file:
        file = json.load(json_file)
        daily_data.append(file)
    
```

```python
# tickers

tickers = []
# 
for file in glob.glob("*.ipynb"):
    if file[0] == '_':
        continue
    tickers.append(file.split('_')[0])
    
```

```python
daily_data_dict = {item['ticker']:item for item in daily_data}
press_data_dict = {item['ticker']:item for item in press_data}


non_press_days = {}
press_changes_val = []

for ticker in tickers:
    if ticker == 'isrg' or ticker == 'lmt' or ticker == 'pfe' or ticker == 'pg' or ticker == 'Untitled.ipynb':
        continue
    cur_data_daily_dates = daily_data_dict[ticker]['data'][0]
    cur_data_daily_price = daily_data_dict[ticker]['data'][1]
    
    cur_data_press = press_data_dict[ticker]['data']
    cur_data_press_dates = [x['date'] for x in cur_data_press]
    
    cur_data_press_changes = [x['1d_change'] for x in cur_data_press]
    
    press_changes_val.extend(cur_data_press_changes)
    
    cur_data_press_dates_day2 = [str(parse(x).date() + datetime.timedelta(days=1)) 
                                 for x in cur_data_press_dates]
    #cur_data_press_dates.extend(cur_data_press_dates_day2)
    main_list = list(np.setdiff1d(cur_data_daily_dates, cur_data_press_dates))
    
    non_press_days_ticker = []
    for i in range(len(cur_data_daily_dates) - 1):
        
        if cur_data_daily_dates[i] in main_list:
            non_press_days_ticker.append([cur_data_daily_dates[i], cur_data_daily_price[i]])

    non_press_days.update({ticker : non_press_days_ticker})            

    
    
```

```python
# Extract list of values for ploting
non_press_days_flat_list = [item[1] for sublist in list(non_press_days.values()) for item in sublist]


sampled_non_press_days = np.random.choice(non_press_days_flat_list, size=len(press_changes_val), replace=True)
```

```python
fig, axs = plt.subplots(1, 3, figsize=(15, 5), sharey=False)

sns.histplot(non_press_days_flat_list, ax=axs[0]).set(xlim=(-20, 20))

sns.histplot(press_changes_val, ax=axs[1]).set(xlim=(-20, 20), 
                                               xlabel='Share price change')

sns.histplot(sampled_non_press_days, ax=axs[2]).set(xlim=(-20, 20))
```

```python
# convert to numpy array to easily calculcate aggregate statistics
non_press_days_flat_list = np.array(non_press_days_flat_list)
press_changes_val = np.array(press_changes_val)

# mean
print(f'Mean for non press changes = {non_press_days_flat_list.mean()}')
print(f'Mean for press changes = {press_changes_val.mean()}')
```

```python
# standard deviation
print(f'Standard deviation for non press changes = {np.array(non_press_days_flat_list).std()}')
print(f'Standard deviation for press changes = {np.array(press_changes_val).std()}')
```

```python
from scipy import stats

# t distribution approximation of confidence intervals for the mean
mean_press_days_confidence_interval = stats.t.interval(0.95, len(press_changes_val)-1, 
                       loc=np.mean(press_changes_val), scale=stats.sem(press_changes_val))

mean_non_press_days_confidence_interval = stats.t.interval(0.95, len(non_press_days_flat_list)-1, 
                       loc=np.mean(non_press_days_flat_list), scale=stats.sem(non_press_days_flat_list))


print(f'Mean confidence interval for press days {mean_press_days_confidence_interval}')
print(f'Mean confidence interval for non press days {mean_non_press_days_confidence_interval}')
```

```python
# t-test
# standard independent 2 sample t-test 
# assumes normal distribution
single_ttest = stats.ttest_ind(press_changes_val, non_press_days_flat_list, equal_var=True)
print(f'Ttest statistic = {single_ttest.statistic} pvalue = {single_ttest.pvalue}')

# Welch t-test
# Visually variance look to be different and supported by values of standard deviation
# Welch test is performed where the group are assumed to have unequal variances
# Also assumes groups are independant
single_ttest_welch = stats.ttest_ind(press_changes_val, non_press_days_flat_list, equal_var=False)
print(f'Welch statistic = {single_ttest_welch.statistic} pvalue = {single_ttest_welch.pvalue}')

# mannwhitneyu test
# Does not assume a close to normal distribution
# parametric test
single_mannwhitneyu = stats.mannwhitneyu(press_changes_val, non_press_days_flat_list)
print(f'Mannwhitneyu statistic = {single_mannwhitneyu.statistic} pvalue = {single_mannwhitneyu.pvalue}')
```

```python
# Bootstrapped ttest

def compute_statistic_ttest(sample1, sample2):
    # calculates t-test p-value
    _, p_value = stats.ttest_ind(sample1, sample2)
    return p_value

num_bootstrap_samples = 1000  # Number of bootstrap samples
bootstrap_statistics_ttest = []
# resample both lists 1000 times
for _ in range(num_bootstrap_samples):
    bootstrap_sample1 = np.random.choice(non_press_days_flat_list, size=len(non_press_days_flat_list), replace=True)
    bootstrap_sample2 = np.random.choice(press_changes_val, size=len(press_changes_val), replace=True)
    statistic = compute_statistic_ttest(bootstrap_sample1, bootstrap_sample2)
    bootstrap_statistics_ttest.append(statistic)
```

```python
# Bootstrapped welch

def compute_statistic_welch(sample1, sample2):
    # calculates welch test p-value
    _, p_value = stats.ttest_ind(sample1, sample2, equal_var=False)
    return p_value

num_bootstrap_samples = 1000  # Number of bootstrap samples
bootstrap_statistics_welch = []
# resample both lists 1000 times
for _ in range(num_bootstrap_samples):
    bootstrap_sample1 = np.random.choice(non_press_days_flat_list, size=len(non_press_days_flat_list), replace=True)
    bootstrap_sample2 = np.random.choice(press_changes_val, size=len(press_changes_val), replace=True)
    statistic = compute_statistic_welch(bootstrap_sample1, bootstrap_sample2)
    bootstrap_statistics_welch.append(statistic)
```

```python
# Bootstrapped mannwhitney

def compute_statistic_mannwhitneyu(sample1, sample2):
    # calculates mannwhitneyu p-value
    _, p_value = stats.mannwhitneyu(sample1, sample2, alternative='two-sided')
    return p_value

num_bootstrap_samples = 1000  # Number of bootstrap samples
bootstrap_statistics_mannwhitneyu = []
# resample both lists 1000 times
for _ in range(num_bootstrap_samples):
    bootstrap_sample1 = np.random.choice(non_press_days_flat_list, size=len(non_press_days_flat_list), replace=True)
    bootstrap_sample2 = np.random.choice(press_changes_val, size=len(press_changes_val), replace=True)
    statistic = compute_statistic_mannwhitneyu(bootstrap_sample1, bootstrap_sample2)
    bootstrap_statistics_mannwhitneyu.append(statistic)
```

```python
confidence_interval_mannwhitneyu = np.percentile(bootstrap_statistics_mannwhitneyu, [2.5, 97.5])
print("95% Confidence Interval mannwhitneyu:", confidence_interval_mannwhitneyu)

confidence_interval_ttest = np.percentile(bootstrap_statistics_ttest, [2.5, 97.5])
print("95% Confidence Interval ttest:", confidence_interval_ttest)

confidence_interval_welch = np.percentile(bootstrap_statistics_welch, [2.5, 97.5])
print("95% Confidence Interval welch:", confidence_interval_welch)
```
